# PREPROCESSING

```python
symptom_columns = [col for col in df.columns if col.lower().startswith("symptom_")]

symptom_data = df[symptom_columns].values.tolist()
```

```python
symptom_columns = [col for col in df.columns if col.lower().startswith("symptom_")]

def clean_value(x):
    if pd.isna(x):
        return None
    x = str(x)
    x = x.strip()                 # hapus spasi depan/belakang
    x = x.lower()                 # lowercase
    x = x.replace(" ", "_")       # spasi jadi underscore
    return x if x != "" else None # buang string kosong

for col in symptom_columns:
    df[col] = df[col].apply(clean_value)
```

```python
symptom_data = []

for _, row in df[symptom_columns].iterrows():
    symptoms = [s for s in row if s is not None]
    symptom_data.append(symptoms)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
X = mlb.fit_transform(symptom_data)

feature_names = mlb.classes_
print("Total clean symptoms:", len(feature_names))
```

```python
y = df["Disease"]
```

```python
print("Shape X:", X.shape)
print("Shape y:", y.shape)
```

```python
X_df = pd.DataFrame(X, columns=feature_names)
X_df
```

# MODELLING

```python
import numpy as np

X_aug = X.copy()

for i in range(len(X_aug)):
    ones_idx = np.where(X_aug[i] == 1)[0]
    if len(ones_idx) > 1:
        drop_idx = np.random.choice(ones_idx, size=1, replace=False)
        X_aug[i][drop_idx] = 0
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_aug, y,
    test_size=0.2,
    random_state=42,
    stratify=y  # penting untuk multiclass
)
```

```python
# from sklearn.linear_model import LogisticRegression

# model = LogisticRegression(max_iter=500)
# model.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import numpy as np

probs = model.predict_proba(X_test)

# contoh lihat 1 sample
sample_probs = probs[0]

top3_idx = np.argsort(sample_probs)[-3:][::-1]

for idx in top3_idx:
    print(model.classes_[idx], sample_probs[idx])
```

# TEST

```python
import numpy as np
import joblib

def predict_disease(user_symptoms):
    model = joblib.load("disease_model.pkl")
    mlb = joblib.load("symptom_encoder.pkl")

    user_vector = mlb.transform([user_symptoms])

    probs = model.predict_proba(user_vector)[0]
    top_idx = np.argsort(probs)[-3:][::-1]

    results = []
    for idx in top_idx:
        results.append({
            "disease": model.classes_[idx],
            "probability": float(probs[idx])
        })

    return results
```

```python
predict_disease(["high_fever", "cough", "headache", "fatigue"])
```
